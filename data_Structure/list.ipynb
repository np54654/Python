# List

#### List are one of the type of data structures which are used for storing the data in the structured format where you can store data of any data type and that it is one of the data structure so we can have the access of the data as well that we are going to stroe into the list and we can apply CRUD operations as well i.e. Create, Read, Update, Delete operations as well.

```python
lists=["Neeraj", "Satyarth", "Anshul", 1,2,3,6.2,True]
```

```python
print(lists)
```

#### List are ordered and are a multiple Data Structure

## List Indexing

```python
## By using the proper indexing value we can access the particular element of the list
## indexing in the list starts from 0 and ends at n-1 where n is the number of elements in the list

print(lists[0]) ## This will print the first element of the list
print(lists[1]) ## This will print the second element of the list
```

```python
print(len(lists)) ## This will print the length of the list
```

```python
lists=["Mango", "Banana", "Apple", "Grapes"]
```

```python
# Negative indexing in the list starts from -1 and ends at -n where n is the number of elements in the list
print(lists[-1]) ## This will print the last element of the list
```

```python
print(len(lists)-1) ## This will print the index of the last element of the list
```

```python
list1=len(lists)-1
print(lists[list1]) ## This will print the last element of the list
```

```python
lists[:]
```

#### List Slicing

List slicing can help us with retriving a specific data points from our list.

```python
lists[1:3] ## This will print the elements from index 1 to index 2 (3-1) of the list
```

#### Append The List or add to the list.

```python
# To add the element at the end of the list we can use the append() method
lists.append("Mango") ## This will add the element "Mango" at the end of the list
print(lists)
```

#### Reverse the list.

```python
lists[::-1] ## This will print the whole list in reverse order
```

```python
lists.reverse() ## This will reverse the list permanently
print(lists)
```

```python
lists.pop() ## This will remove the last element of the list
print(lists)
```

```python
lists.index("Apple") ## This will return the index of the first occurrence of the element "Apple" in the list
```

```python
lists[1:] ## This will print the elements from index 1 to the end of the list
```

```python
lists[:-1] ## This will print the elements from the beginning of the list to the second last element of the list
```

# Lists are Mutable

#### By mentioning that lists are mutable we mean to say that we can replace the values present at a specific location inside our list

```python
#To update the element of the list we can use the indexing value and assign the new value to it
lists[0]="Pineapple"
print(lists)
```
