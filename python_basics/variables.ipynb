## input Function: this function can take input from the user side and then it can store that input into an variable which is defined.

```python
Name = input("Enter your name: ")
print("My name is " + Name) 
```

```python
age = int(input("Enter your age: ")) #Here we are converting the input string to integer using int() function
if age > 18:
    print("You are eligible for driving license")   
else:
    print("You are not eligible for driving license")
print(type(age))
```
