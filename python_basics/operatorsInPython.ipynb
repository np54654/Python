# Operators

## 1. Arithmetic Operators:

'+' Addition
'-' Substraction
'*' Multiplication
'/' Division
'//' Floor
'a**b' Exponent
'%' Reminder (Modulus)

## 2. Comparision Operator

'=' Assigning
'>' greater than
'<' Less than
'>=' Greater than or equals to
'<=' Less than or equal to 
'==' Equals to
'!=' Not equals to

## 3. Logical Operator

'and' it checks if both the operations or all the operations are true or not.
'or' it checks if any one operation it true or not.
'not' inverse the result of the given operation or negation.

```python
a=14
b=3
c=a%b #modulo operator: it gives the remainder after dividing a by b
print(c)
d=a//b #floor division: it gives the quotient without the remainder
print(d)
```

```python
a=2
b=3
c=a**b #exponentiation operator: it gives the result of a raised to the power of b
print(c)
```

```python
# Calculator

a=float(input("Enter first number: ")) #input() function takes input from the user and returns it as a string. We need to convert it to float using float() function.
b=float(input("Enter second number: ")) 
operator=input("Enter operator (+, -, *, /): ") #input() function takes input from the user and returns it as a string. We can directly use it as operator.
if operator=='+': #if the operator is +, then we perform addition
    result=a+b  #we can use the + operator to add two numbers
elif operator=='-':#if the operator is -, then we perform subtraction
    result=a-b
elif operator=='*':
    result=a*b  
elif operator=='/':
    result=a/b
else:
    result="Invalid operator"
print("Result:", result)
```
