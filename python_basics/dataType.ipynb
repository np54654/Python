# Data Types

> Data types are the classification of the data that tells the compiler or the interpretor that what kind of values we can store into a variable and how much amount of the memory will be consumed to store the data and the kind of operation it can perform

```python
age=input("Enter the age:")
```

```python
age
```

```python
type(age)
```

```python
age2=40
```

```python
age+age2
# The above code will give an error because age is a string and age2 is an integer. We need to convert age to an integer before adding it to age2.
```

```python
# To convert age to an integer, we can use the int() function.
age=int(age)
```

```python
age+age2
```

```python
# Or we can use int() function directly in the addition operation.
int(age)+age2
# Or we can use the int() function to convert age to an integer and then add it to age2, in the first line of code.
age=int(input("Enter the age:"))
```

```python
# We can have another example for String data type.
age2='hello'
age=input("Enter the age:")
```

```python
print(age2)
print(age)
```

```python
age+age2
# Now here the 2 values are of string data type, so we can concatenate them. The output will be the combination of both strings.    
```
